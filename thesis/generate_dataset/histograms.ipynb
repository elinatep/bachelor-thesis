```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path1 = '/datasets/eteplygina/filtered_positive_data.csv'
data1 = pd.read_csv(file_path1)
file_path2 = '/datasets/eteplygina/filtered_negative_data.csv'
data2 = pd.read_csv(file_path2)

# Drop duplicate entries for each patient
data_unique1 = data1.drop_duplicates(subset=['id'])
data_unique2 = data2.drop_duplicates(subset=['id'])

# Plot histogram for Age
plt.figure(figsize=(10, 6))
plt.hist(data_unique1['age'], bins=20, edgecolor='black')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.title('Histogram of Age Distribution Among Positive Patients', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()

# Plot histogram for Age
plt.figure(figsize=(10, 6))
plt.hist(data_unique2['age'], bins=20, edgecolor='black')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.title('Histogram of Age Distribution Among Negative Patients', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Plot histogram for Gender
plt.figure(figsize=(10, 6))
data_unique1['gender'].value_counts().plot(kind='bar', edgecolor='black')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.title('Histogram of Gender Distribution Among Positive Patients', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()

# Plot histogram for Gender
plt.figure(figsize=(10, 6))
data_unique2['gender'].value_counts().plot(kind='bar', edgecolor='black')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.title('Histogram of Gender Distribution Among Negative Patients', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Plot histogram for Ethnicity
plt.figure(figsize=(10, 6))
data_unique1['ethnicity'].value_counts().plot(kind='bar', edgecolor='black')
plt.xlabel('Ethnicity', fontsize=14)
plt.ylabel('Number of Patients',fontsize=14)
plt.title('Histogram of Ethnicity Distribution Among Positive Patients', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()

# Plot histogram for Ethnicity
plt.figure(figsize=(10, 6))
data_unique2['ethnicity'].value_counts().plot(kind='bar', edgecolor='black')
plt.xlabel('Ethnicity', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.title('Histogram of Ethnicity Distribution Among Negative Patients', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/datasets/eteplygina/filtered_positive_data.csv'
data = pd.read_csv(file_path)

# Convert 'time_to_shock' to hours
data['time_to_shock'] = pd.to_timedelta(data['time_to_shock']).dt.total_seconds() / 3600

# Drop duplicate entries for each patient
data_unique = data.drop_duplicates(subset=['id'])

# Plot histogram
plt.figure(figsize=(12, 7))
plt.hist(data_unique['time_to_shock'], bins=20, edgecolor='black')
plt.xlabel('Time to Shock (hours)', fontsize=14)
plt.ylabel('Number of Patients', fontsize=14)
plt.title('Histogram of Time to Shock in Hours', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)
plt.show()
```
